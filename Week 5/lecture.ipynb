```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)  # exactly 10 points from start point to end point
 # x = np.arrrang(0, 1, 0.1)  # steps of 0.1 


c = lambda n: np.cos(n*2*np.pi*x)
s = lambda n: np.sin(n*x)
```

```python
plt.plot(x, c(1))
```

```python
fig, ax = plt.subplots(2, 4, sharex = True, sharey=True)


for j in range(4):
    ax[0,j].plot(x, c(j), color = 'blue', lw=2.5)
    ax[0,j].set_title(f'cos({2*j}*pi*x)')
 ### ax[0,1].plot(x, c(1), color = 'blue', lw=2.5)
 #### ax[0,2].plot(x, c(2), color = 'blue', lw=2.5)
```
